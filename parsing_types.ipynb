```python
import ujson
import gzip
import matplotlib
import pandas
import os
import re
import time
import pprint
```

```python
def parse_type(type, num_examples = 5):
    examples = []
    num_found = 0
    with gzip.open('2015-10-01-15.json.gz') as f:
        for i, line in enumerate(f):
            json_data = ujson.loads(line)
            if (json_data['type'] == type):
                examples.append(json_data)
                num_found += 1
                if (num_examples < num_found):
                    f.close()
                    return(examples)
        f.close()
    return(examples)
def parse_notype(num_examples = 5):
    examples = []
    num_found = 0
    with gzip.open('2015-10-01-15.json.gz') as f:
        for i, line in enumerate(f):
            json_data = ujson.loads(line)
            if not json_data['type']:
                examples.append(json_data)
                num_found += 1
                if (num_examples < num_found):
                    f.close()
                    return(examples)
        f.close()
    return(examples)

def parse_language_event(num_examples = 5):
    examples = []
    num_found = 0
    regexp = re.compile(r'language')
    with gzip.open('2015-10-01-15.json.gz') as f:
        for i, line in enumerate(f):
            if regexp.search(line.decode()) is not None:
                json_data = ujson.loads(line)
                examples.append(json_data)
                num_found += 1
                if (num_examples < num_found):
                    f.close()
                    return(examples)
        f.close()
    return(examples)
```

```python
language_events = []
for event in parse_language_event(9999999):
    language_events.append(ujson.dumps(event))
print("done")
```

```python

types = []
for i, line in enumerate(language_events):
    json_data = ujson.loads(line)
    types.append(json_data['type'])
```

```python
pandas.DataFrame({"types":types}).groupby("types").size().order(ascending=False)
```
